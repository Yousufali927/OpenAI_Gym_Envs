```python
import os 
import gym
from stable_baselines3 import DQN
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
# making the environment using the OpenAI Gym envs
```

```python
env_name = 'MountainCar-v0'
env = gym.make(env_name, render_mode='human')
```

```python
# giving the log path to store the results/data related to the training which we can review in tensorboard
```

```python
log_path = os.path.join('500k_DQN_Training_MC', 'Logs')
```

### Defining and Training the model for 100k timesteps

```python
env = DummyVecEnv([lambda: env])
model = DQN('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=100000)
```

### Saving the model

```python
DQN_Path = os.path.join('DQN_MCar', 'Saved Models', '40k_Mountain_Car') 
```

```python
model.save(DQN_Path)
```

### Testing the model

```python
test_env = gym.make(env_name, render_mode='human')
```

```python
# Testing for 5 epochs
episodes = 5
for episode in range(1,episodes+1):
    obs, _ = test_env.reset()
    done=False
    score =0
    
    while not done:
        action, _ = model.predict(obs, deterministic=True)
        obs, reward, terminated, truncated, _ = test_env.step(action)
        done = terminated or truncated
        score += reward
        test_env.render()
        print('Episode:{} Score:{}'.format(episode, score))

test_env.close()
```
