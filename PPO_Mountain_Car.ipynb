```python
import os 
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
env_name = 'MountainCar-v0'
env = gym.make(env_name, render_mode='human')
```

```python
log_path = os.path.join('500k_PPO_Training_MC', 'Logs')
```

```python
env = DummyVecEnv([lambda: env])
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
## Execute this 
## I have executed it in another ipynb file so I will just load that model here
model.learn(total_timesteps=100000)
```

### saving the model

```python
test_env = gym.make(env_name, render_mode='human')
```

```python
PPO_Path = os.path.join('Training_MC', 'Saved Models', '500k_Mountain_Car') 
```

```python
# execute this after trainning to save your model
model.save(PPO_Path)
```

```python
model = PPO.load(PPO_Path, env=test_env)
```

### Testing the model

```python
# Testing for 5 epochs
episodes = 5
for episode in range(1,episodes+1):
    obs, _ = test_env.reset()
    done=False
    score =0
    
    while not done:
        action, _ = model.predict(obs, deterministic=True)
        obs, reward, terminated, truncated, _ = test_env.step(action)
        done = terminated or truncated
        score += reward
        test_env.render()
        print('Episode:{} Score:{}'.format(episode, score))

test_env.close()
```
